```python
import numpy as np
import matplotlib.pyplot as plt


# Dinamica del capital

# Parametros
epsilon = 0.5  
K_o = 4.0  
K_0 = 3.0 

# Función para la solución propuesta
def solucion_propuesta(t):
    return (K_0 - K_o) * (1 / (1 + epsilon)) ** t + K_o

# Configurar el número de pasos
num_periodos = 20

# Calcular los valores de K a lo largo de los pasos utilizando la solución propuesta
K_valores_propuestos = [solucion_propuesta(t) for t in range(num_periodos)]

# Crear un array de valores de tiempo (t)
tiempo = np.arange(0, num_periodos)

# Graficar los resultados
plt.plot(tiempo, K_valores_propuestos, marker='o', linestyle='-', color='b', label='Trayectoria')
plt.xlabel('Periodos (t)')
plt.ylabel('K(t)')
plt.title('Dinámica del capital')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Dinamica de la inversion

# Parametros
epsilon = 0.1
K_o = 5.0  
K_0 = 3.0 

# Función para la solución propuesta
def solucion_propuesta(t):
    return (epsilon/(1+epsilon))*(K_o - K_0) * (1 / (1 + epsilon)) ** t

# Configurar el número de pasos
num_periodos = 20

# Calcular los valores de K a lo largo de los pasos utilizando la solución propuesta
K_valores_propuestos = [solucion_propuesta(t) for t in range(num_periodos)]

# Crear un array de valores de tiempo (t)
tiempo = np.arange(0, num_periodos)

# Graficar los resultados
plt.plot(tiempo, K_valores_propuestos, marker='o', linestyle='-', color='b', label='Trayectoria')
plt.xlabel('Periodos (t)')
plt.ylabel('I(t)')
plt.title('Dinámica de la inversión')
plt.grid(True)
plt.legend()
plt.show()
```
